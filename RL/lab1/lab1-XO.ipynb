```python
import numpy as np
import gym
from stable_baselines3 import PPO
import math
import itertools
import datetime

import gym
from lab1 import Player2, XO_Env
from stable_baselines3.common.env_checker import check_env
from stable_baselines3 import PPO

from gym.envs.registration import register
register(
    id='XO_Env-v0',
    entry_point='lab1:XO_Env',
    max_episode_steps=20
)
```

```python
p2 = Player2()
learning_env = gym.make('XO_Env-v0',player2=p2)
#TSTEPS = int(input("TSTEPS: "))
learning_env.set_human(False)
learning_env.set_display(False)
```

```python
learning_env.display
```

```python
model = PPO('MlpPolicy', learning_env,learning_rate=0.0003,gamma=0.999, verbose=True)
model.learn(total_timesteps=100000)
model.save(f'agents/agent1')
```

```python
epis = 20
testing_env = gym.make('XO_Env-v0', player2=p2)
obs = testing_env.reset()
testing_env.set_human(True)
testing_env.set_display(True)
```

```python
testing_env.display
```

```python
for epi in range(epis):
    done = False
    while not done:
        action, _state = model.predict(obs, deterministic=True)
        obs, reward, done, info = testing_env.step(action)
        #testing_env.render()
        
        if done:
            print('Done!', reward)
    obs = testing_env.reset()

    print(f'Game #{epi} over\n\n\n')
```
